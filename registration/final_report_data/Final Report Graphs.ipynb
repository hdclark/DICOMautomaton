```python
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_csv("synthetic/rigid/rigid_stats.csv")
df2 = pd.read_csv("synthetic/affine/affine_stats.csv")
df3 = pd.read_csv("synthetic/nonrigid/nonrigid_stats.csv")

fig, ax3 = plt.subplots()

ax3.set_xlabel('Time(s)')
ax3.set_ylabel('RMS Error')
# ax3.set_xscale("log")
ax3.plot(df1["time"], df1["error"])
ax3.plot(df2["time"], df2["error"])
ax3.plot(df3["time"], df3["error"])
ax3.tick_params(axis='y')
ax3.set_title("RMS Error v.s. Time for Rigid, Affine, and Non-Rigid Algorithms")

ax3.legend(["rigid", "affine", "nonrigid"])
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
